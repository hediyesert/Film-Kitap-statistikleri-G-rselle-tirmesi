```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veri setinizin dosya adını buraya yazın
dosya_adi = 'movies_dataset.csv'  # Örnek dosya adı

try:
    # Veri setini oku (Dosyanın dizinini kontrol etmeniz gerekebilir)
    df = pd.read_csv(dosya_adi)
    print("Veri seti başarıyla yüklendi.")
except FileNotFoundError:
    print(f"HATA: '{dosya_adi}' dosyası bulunamadı. Dosya adını kontrol edin.")
    exit()

# Verinin ilk 5 satırını inceleyin (Sütun isimlerini anlamak için)
print("\nVerinin İlk 5 Satırı:")
print(df.head())
```

```python
# A) Eksik Değerleri Kontrol Etme
print("\nEksik Veri Sayısı:")
print(df.isnull().sum())

# B) Puan ve Yıl Sütunlarını Kontrol Etme
# Eksik Puan değerlerini veri setinin ortalaması ile dolduralım (Basit bir temizlik)
if 'rating' in df.columns and df['rating'].isnull().any():
    ortalama_puan = df['rating'].mean()
    df['rating'].fillna(ortalama_puan, inplace=True)
    print(f"\n'rating' sütunundaki eksik değerler ortalama puan ({ortalama_puan:.2f}) ile dolduruldu.")

# C) Yıl Sütununu Düzenleme (Eğer yıl metin olarak geldiyse)
# Eğer yıl sütununuz metin içinde ise (örneğin '2020 (I)'), sadece sayıyı almalısınız.
if 'year' in df.columns:
    # Eğer 'year' sütunu string ise ve yıl parantez içinde ise
    df['year'] = df['year'].astype(str).str.extract('(\d{4})').astype(float)
    # Artık sayısal olduğu için, NaN (boş) olanları temizleyebiliriz (ya da silebiliriz)
    df.dropna(subset=['year'], inplace=True)
    df['year'] = df['year'].astype(int)
```

```python
plt.figure(figsize=(12, 6))

# Matplotlib'in temel scatter fonksiyonu: plt.scatter()
plt.scatter(x=df['Released_Year'], y=df['IMDB_Rating']) 

plt.title('Film Puanlarının Yıla Göre Dağılımı')
plt.xlabel('Yapım Yılı')
plt.ylabel('IMDb Puanı (Rating)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Seaborn ile Histogram çizimi
sns.histplot(df['IMDB_Rating'], bins=20, kde=True) # kde=True, bir yoğunluk çizgisi ekler

plt.title('Film Puanlarının Frekans Dağılımı')
plt.xlabel('Puan')
plt.ylabel('Film Sayısı (Frekans)')
plt.show()
```

```python
# En yüksek puana sahip ilk 10 yönetmeni bulma
yonetmen_siralamasi = df.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(14, 7))
sns.barplot(x=yonetmen_siralamasi.index, y=yonetmen_siralamasi.values, palette="viridis")

plt.title('En Yüksek Ortalama Puanlı İlk 10 Yönetmen')
plt.xlabel('Yönetmen Adı')
plt.ylabel('Ortalama Puan')
plt.xticks(rotation=45, ha='right') # X eksenindeki isimleri yatay yapma
plt.tight_layout() # Başlıkların kesilmemesi için
plt.show()
```
